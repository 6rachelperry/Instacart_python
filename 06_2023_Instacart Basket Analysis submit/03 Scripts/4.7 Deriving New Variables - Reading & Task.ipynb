```python
#making a subsest of first 1million rows
df_sub = df_orders_prods_merged[:1000000]
```

```python
# making a price range function using IF 
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range product'
  else: return 'Not enough data'
```

```python
#applying the new function
df_sub['price_range'] = df_sub.apply(price_label, axis=1)
```

```python
#value counts
df_sub['price_range'].value_counts(dropna = False)
```

```python
# using a loc function to make a price range statement
df_sub.loc[df_sub['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# using a loc function to make a price range statement
df_sub.loc[(df_sub['prices'] <= 15) & (df_sub['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```
