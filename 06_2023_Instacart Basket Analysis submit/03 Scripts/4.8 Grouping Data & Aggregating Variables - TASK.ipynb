## 3.2 Grouping loyalty customers and checking aggregations

```python
df_ords_prods_merge.head()
```

```python
#4 creating loyalty flag column

df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#checking freq counts
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merge.head()
```

```python
#5 seeing if there is a spending difference based on type of customer
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

## it looks like loyal customers find a way to spend less on the items they are purchasing

## 3.3 Aggregating average prices 

```python
#6 grouped by 'user_id' and 'prices' transformed into mean of 'prices' making new column avg prices

df_ords_prods_merge['avg_prices'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge.head()
```

## 3.4 grouping a spending flag

```python
#6 creating spending flag column
df_ords_prods_merge.loc[df_ords_prods_merge['avg_prices'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['avg_prices'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
#checking freq counts
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merge.head()
```

## 3.5 Aggregating median order frequency

```python
#7 grouped by 'user_id' and 'days since prior' transformed into median order freq making new column 

df_ords_prods_merge['median_order_freq'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge.shape
```

## 3.6 grouing a frequency customer order flag

```python
#7 creating a freq customer flag
df_ords_prods_merge.loc[df_ords_prods_merge['median_order_freq'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_order_freq'] <= 20) & (df_ords_prods_merge['median_order_freq'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_order_freq'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
#checking freq counts
df_ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
